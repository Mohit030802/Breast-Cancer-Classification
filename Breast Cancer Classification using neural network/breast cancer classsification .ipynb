## separating between features and target

```python
X=breast_cancer_df.drop('label',axis=1)
Y=breast_cancer_df['label']
```

```python
X.head()
```

```python
Y.head()
```

## Training and Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

## Standardize the data

```python
from sklearn.preprocessing  import StandardScaler
```

```python
scalar=StandardScaler()
X_train_standard=scalar.fit_transform(X_train)
X_test_standard=scalar.transform(X_test)
```

```python
print(X_train_standard)
```

## Building the neural network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

### setting up the layers of my neural network

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')
                       ])
```

```python
# compling the neural networks
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# training the neural network
history=model.fit(X_train_standard,Y_train,validation_split=0.1,epochs=10)
```

## Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])

plt.plot(history.history['val_accuracy'])


plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])

plt.plot(history.history['val_loss'])


plt.title('model loss')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='upper right')
```

## accuracy on test data

```python
loss,accuracy=model.evaluate(X_test_standard,Y_test)
```

## Precdictive system

```python
Y_pred=model.predict(X_test_standard)
```

```python
Y_pred.shape
```

```python
Y_pred
# basically model.predict gives the prediction probabilty
# here the first value repersents the probabilty of 0 and other repersents the probability of 1
```

## converting the prediction probabilty to labels

```python
Y_pred_labels=[np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

## Building the predictive system

```python
input_data=[X_test_standard[8]]
input_data_toarray=np.asarray(input_data)
input_data_reshaped=input_data_toarray.reshape(1,-1)
# standardize the data
input_data_std=scalar.transform(input_data_reshaped)
prediction=model.predict(input_data_std)
print(prediction)
prediction_label=[np.argmax(prediction)]
print(prediction_label)
if(prediction_label[0]==0):
    print('Maligant')
else:
    print('Benign')
```
